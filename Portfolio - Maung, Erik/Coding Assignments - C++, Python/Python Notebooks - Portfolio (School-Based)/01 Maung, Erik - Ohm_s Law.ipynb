```python
import numpy as np
read_in_array = np.loadtxt('/content/drive/MyDrive/2021-2022/22Spring/4BL/Reference Files/01voltagedata.txt',delimiter=',')
```

```python
#conversion factor = 0.0049
voltage2k = 0.0049 * read_in_array[:,0]
```

```python
#calculate current = voltage / resistance
#resistance = 2k Ohms
current2k = voltage2k / 2000
```

```python
import matplotlib.pyplot as plt
plt.scatter(voltage2k,current2k)
plt.xlabel('Voltage (Volts)')
plt.ylabel('Current (Amps)')
```
