```python
# Read in 3 color LED data, find the right source in your google drive
# The columns in your dataset were separated with a semicolon, we set that as the delimiter
led3_data = np.loadtxt('/content/gdrive/MyDrive/2021-2022/22Winter/4AL/Python References/1.25 Multicolor/3color_reaction.txt', delimiter=";")
```

Separate out the two columns, we only want to work with the reaction time.

```python
# Extract only the reaction times from the data
led3reaction = led3_data[:,1]
print(led3reaction)
```

You can find the mean and standard deviation of your reaction time values using np.mean() and np.std().

```python
# Find the mean
# Print the mean
led3mean = np.mean(led3reaction)
print('The mean of all 3-color reaction times is ' + str(np.around(led3mean, 3)) + ' milliseconds.')
# Find the standard deviation
# Print the standard deviation
led3std = np.std(led3reaction)
print('The st. dev. of all 3-color reaction times is ' + str(np.around(led3std, 3)) + ' milliseconds.')
```

# Histograms

```python
# Plot the first histogram from 3 color LED data
plt.hist(led3reaction, bins = 10, alpha = 0.8, color = 'lightblue', label = '3-Color')
# bins is the number of bins and alpha sets transparency to your histogram


# Plot the second histogram from 2 color LED data - has to be in the same plot window
plt.hist(led2reaction, bins = 10, alpha = 0.5, color = 'green', label = '2-Color')

# Plot the thrid histogram from the LED data of Lab 1E - has to be in the same plot window
# Choose different colors for all the histograms
plt.hist(led1reaction, bins = 10, alpha = 0.8, color = 'pink', label = '1-Color')

# Provide axes labels
plt.xlabel('Reaction (milliseconds)')
plt.ylabel('Frequency')

# Add legends - you will need to add "labels" to your histograms and use plt.legend()
plt.legend()
```
